```python
import spacy
import glob
import os
import json
import pandas as pd
from spacy.tokens import DocBin
from spacy import displacy
```

```python
def add(self, data):
        # Add something to the component's data
        self.data.append(data)

def to_disk(self, path, exclude=tuple()):    # This will receive the directory path + /my_component   
    data_path = path / "data.json"        
    with data_path.open("w", encoding="utf8") as f:            
        f.write(json.dumps(self.data))

def from_disk(self, path, exclude=tuple()):        # This will receive the directory path + /my_component       
    data_path = path / "data.json"        
    with data_path.open("r", encoding="utf8") as f:            
        self.data = json.loads(f)        
    return self
```

```python
def checkContinuous(nums):
   if len(nums) < 1:
      return False
   min_val = min(nums)
   max_val = max(nums)
   if max_val - min_val + 1 == len(nums):
      for i in range(len(nums)):
         if nums[i] < 0:
            j = -nums[i] - min_val
         else:
            j = nums[i] - min_val
            if nums[j] > 0:
               nums[j] = -nums[j]
            else:
               return False
      return True
   return False
```

```python
def checkSingleton(token):
    if token.children == 0:
        return True
    else:
        return False
```

```python
#Notes (doc attributes):
#text, left_edge, right_edge, ent_type_, ent_iob_, lemma_, n_lefts
def getMentionsDataframe(doc):
    df_preparation_list = []
    for i, row in df.iterrows():
        doc = row["doc"]
        filename = row["filename"]
        description = row["description"]
        doc_id = row["source_domain"]
        for j, sentence in enumerate(doc.sents):
            for token in sentence:
                if (token.dep_ == "root" or token.dep_ == "pobj" ) and not token.is_punct:
                    tokens = []
                    tokens_str = ""
                    tokens_number = []
                    for t in token.head.subtree:
                        tokens.append(t)
                        tokens_str = tokens_str + " " + t.text
                        #if len(tokens) > 10:
                        #    tokens_str = ""
                        #    nearby_tokens = doc[token.i-7:token.i]
                        #    for n in nearby_tokens:
                        #        if not n.is_punct:
                        #            tokens_str = tokens_str + " " + n.text
                        tokens_number.append(t.i)
                        
                    sent_id = j
                    coref_chain = token.ent_type_[:3] + str(token.head.norm)
                    mention_type = token.ent_type_
                    is_continuous = checkContinuous(tokens_number[:])
                    is_singleton = checkSingleton(token)
                    mention_id = doc_id+"_"+str(sent_id)+"_"+str(tokens_number[0])
                    
                    #add to the list
                    df_preparation_list.append([coref_chain, doc_id, is_continuous, is_singleton, mention_id, mention_type[:3], mention_type, "score", sent_id, tokens_number, tokens_str[1:], filename, "coref_type", description, token.head.text, token.head.i])
                    
    mentions_df = pd.DataFrame(df_preparation_list, columns=["coref_chain", "doc_id", "is_continuous", "is_singleton", "mention_id", "mention_type", "mention_full_type", "score", "sent_id", "tokens_number", "tokens_str", "topic_id", "coref_type", "description", "head_str", "head_id"])
    #tokens_df.append( columns = df_preparation_list)
    return mentions_df
```

```python

mentions_df = getMentionsDataframe(df)
mentions_df.head(10)  
```
